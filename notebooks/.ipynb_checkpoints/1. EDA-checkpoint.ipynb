### loan_amnt

```python
get_info('loan_amnt')
```

```python
loan_amnt = data['loan_amnt']

bins = [bn for bn in range(0, 41000, 2500)]
x_labels = [str(int(bn/1_000)) + 'K' for bn in bins[::2]]

plt.hist(loan_amnt, bins=bins, edgecolor='black')

loan_amnt_median = loan_amnt.median()
color = '#fc4f30'

plt.axvline(loan_amnt_median, color=color, label='Median Amount', linewidth=2)

plt.legend()

plt.title('Amount of Loans')
plt.xlabel('Amount')
plt.ylabel('Number of Loans')

plt.xticks(ticks=bins[::2], labels=x_labels)

plt.tight_layout()

#plt.savefig('plots/loan_amnt_hist.png')

plt.show()
```

### Correlation Matrix

```python
data.corr()
```

```python
# Compute the correlation matrix
corr = data.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, annot=True, vmax=1, vmin=-1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

# fix for mpl bug that cuts off top/bottom of seaborn viz
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values

plt.title('Correlation Matrix')

plt.tight_layout()

#plt.savefig('plots/corr_map.png')

plt.show()
```

# Missing Values

```python
def get_percent_missing(col):
    missing_vals = data[col].isna().sum()
    total_vals = data.shape[0]
    percent = round(100*missing_vals/total_vals,2)
    return (str(percent) + '%')
```

```python
cols = [col for col in data.columns]
percent = [get_percent_missing(col) for col in data.columns]

missing_df = pd.DataFrame({'column': cols, 'percentage': percent})
```
